Import libraries

```python
import cv2
import matplotlib.pyplot as plt
from image_analysis import process_images
```

Function for visualization satellite image

```python
def load_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        raise ValueError(f"Failed to load image: {image_path}")
    return image

def visualize_image(image_path):
    try:
        # Load the image from the specified path
        image = load_image(image_path)
        
        # Display the image using matplotlib
        plt.figure(figsize=(10, 10))
        plt.imshow(image, cmap='gray')
        plt.title(f"Visualizing: {image_path.split('/')[-1]}")
        plt.axis('off')  # Hide axis
        plt.show()

    except Exception as e:
        print("Error visualizing image")
```

Accessing an image from one folder, yet ensuring it differs from the previous onee

```python
file_path1 = "database/S2A_MSIL1C_20180919T083621_N0206_R064_T36UXA_20180919T105540.SAFE/GRANULE/L1C_T36UXA_A016935_20180919T084300/IMG_DATA/T36UXA_20180919T083621_B04.jp2"
file_path2 = "database/S2A_MSIL1C_20180919T083621_N0206_R064_T36UXA_20180919T105540.SAFE/GRANULE/L1C_T36UXA_A016935_20180919T084300/IMG_DATA/T36UXA_20180919T083621_B07.jp2"
visualize_image(file_path1)
visualize_image(file_path2)
result_image  = process_images(file_path1, file_path2)


plt.figure(figsize=(12, 6))
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
